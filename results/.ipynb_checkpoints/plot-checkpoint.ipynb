```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
resnet_file = './resnet50.csv'
resnet_finetuned_file = './resnet50_finetuned.csv'
```

```python
resnet_result = pd.read_csv(resnet_file)
resnet_finetuned_result = pd.read_csv(resnet_finetuned_file)
```

```python
resnet_result
```

```python
resnet_finetuned_result
```

```python
x = resnet_result.epoch.to_list()
y1 = resnet_result.acc.to_list()
y2 = resnet_result.loss.to_list()
y3 = resnet_result.val_acc.to_list()
y4 = resnet_result.val_loss.to_list()
```

```python
x_f = resnet_finetuned_result.epoch.to_list()
y1_f = resnet_finetuned_result.acc.to_list()
y2_f = resnet_finetuned_result.loss.to_list()
y3_f = resnet_finetuned_result.val_acc.to_list()
y4_f = resnet_finetuned_result.val_loss.to_list()
```

```python
plt.plot(x, y1, 'r', label='train_acc')
plt.plot(x, y3, 'b', label='val_acc')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(x, y2, 'r', label='train_loss')
plt.plot(x, y4, 'b', label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(x_f, y1_f, 'r', label='train_acc')
plt.plot(x_f, y3_f, 'b', label='val_acc')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(x_f, y2_f, 'r', label='train_loss')
plt.plot(x_f, y4_f, 'b', label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
